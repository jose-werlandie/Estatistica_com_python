```python
#grafico para inlustrar os resultados 
x1 = (41593/100)
x2 = (24588)
x3 = (17005)
y1 = x2/x1
y2 = x3/x1

def plota_pizza_1():
   labels = ['Homens', 'Mulheres'] 
   titulos = [y1, y2]
   cores = ['lightblue', 'red']
   explode = (0.1, 0)  # somente explode primeiro pedaço
   total = sum(titulos)
   plt.figure(figsize=(6, 6))
   plt.pie(titulos, explode=explode, labels=labels, colors=cores, autopct=lambda p: '{:.0f}'.format(p * total / 100), shadow=True, startangle=90,)

# Determina que as proporções sejam iguais ('equal') de modo a desenhar o círculo

plota_pizza_1()

plt.show()
```

```python
x = dataset.groupby(u'Segmento de Mercado')[u'Segmento de Mercado'].count()
x = x[x > 20]
df = dataset[dataset[u'Segmento de Mercado'].isin(x.keys())]
df = df[df[u'Avaliação Reclamação']==u'Não Resolvida']
x1 = df.groupby(u'Segmento de Mercado')[u'Segmento de Mercado'].count()
print ("Reclamações não resolvidas por Segmento de Mercado")
print (x1)
print ("\n")
```

```python
x = dataset.groupby(u'Segmento de Mercado')[u'Segmento de Mercado'].count()
#Filtrando os segmentos de mercado com mais de 20 reclamações
x = x[x > 20]
df = dataset[dataset[u'Segmento de Mercado'].isin(x.keys())]
df = df[df[u'Avaliação Reclamação']==u'Resolvida']
x1 = df.groupby(u'Segmento de Mercado')[u'Segmento de Mercado'].count()

print ("Reclamações resolvidas por Segmento de Mercado")
print (x1)
print ("\n")
```

```python
df.groupby(u'Segmento de Mercado')[u'Segmento de Mercado'].count().plot.barh(figsize=(6, 6))
```

<p>No gráfico abaixo temos os segmentos de mercados dispostos através de cobranças e contestação. Logo vemos uma maior protuberância no setor de Internet e TV por assinatura seguido por Comercio eletrônico e demais segmentos que aparecem com menos ou quase nenhuma representação gráfica.

```python
segmento = dataset[dataset[u'Grupo Problema']==u'Cobrança / Contestação']
segmento['Segmento de Mercado'].value_counts()
fig, ax = plt.subplots(figsize=(8, 10))
y_pos = np.arange(len(segmento['Segmento de Mercado'].value_counts()))
values = segmento['Segmento de Mercado'].value_counts()
segmentos = segmento['Segmento de Mercado'].unique()
ax.barh(y_pos, values, align='center', color='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(segmentos)
ax.invert_yaxis()
ax.set_xlabel('Reclamacoes')
ax.set_title('Segmento de mercado -  Cobrança / Contestação')
plt.show()
```
