```python
14. Conte o número de valores ausentes em cada coluna do df . Qual coluna tem o número máximo de valores ausentes?

```

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
maximo = df.isnull().sum().max()
maiorErro = ''

for coluna in df.columns:
    erros = df[coluna].isnull().sum()
    print(f"{coluna}: {erros}")
    
    if ( df[coluna].isnull().sum() == maximo ): maiorErro = coluna

print(f"\no que obteve maior quantidade de erros foi: {maiorErro} com {maximo} erros.")
```

16. Como alterar a ordem das colunas de um dataframe? Para isso, respondam as três questões:

a-Em df, troque as colunas 'a' e 'c'
b-Crie uma função genérica para trocar duas colunas, sem codificar os nomes das colunas
c-Classifique as colunas em ordem alfabética inversa, ou seja, colume ’e’ primeiro até a coluna ’a’ last.

```python

#letra A
df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
temp = df['a'].copy()
df['a'] =  df['b']
df['b'] = temp
df
```

```python
#letra B
df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))

# changeColumns([dataframe], 'nomeDaColuna', 'nobeDaOutraColuna')
def changeColumns( data, colA, colB ):
    temp = data[colB].copy()
    data[colB] = data[colA]
    data[colA] = temp

changeColumns( df, 'a', 'b')
df
```

18. A partir de df , filtre o ’Manufacturer’, ’Model’ e ’Type’ para cada 20 linhas, começando em 1º (linha 0)



```python

def groupBy(df, inicio, passo):
    print( df.iloc[inicio:inicio+passo, 0:3], "\n" )
    
    if inicio < len(df):
        return groupBy(df, inicio+passo, passo)
        
groupBy(df, 0, 20)
```

19. Obtenha as duas últimas linhas de df cuja soma da linha é maior que 100

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
print(df)

def test(df):
    lista = []
    
    for i in range(len(df)):
        if ( df.iloc[-i].sum() > 100 ):
            lista.append(df.iloc[-i])

            if len(lista) == 2: 
                return lista

lista = test(df)
print(f"\nindice {lista[0].name} : {lista[0].sum()}\nindice {lista[1].name} : {lista[1].sum()}")
```
