```python
import pandas as pd
import numpy as np
```

```python
Car_Price = {'Company': ['Toyota','Honda','BMW','Audi'],'Price':[23845,17995,135925,71400]}
Car_Horsepower ={'Company':['Toyota','Honda','BMW','Audi'],'horsepower':[141,80,182,160]}
```

```python
Car_Dp = pd.DataFrame(Car_Price)       # data frame car price
Car_Dh = pd.DataFrame(Car_Horsepower)  # data frame car horsepower
```

```python
Car_Dp
```

```python
Car_Dh
```

```python
#Series nada mais é que um array de 1 dimensão.
#partindo dessa afirmação optei por trabalhar com cada coluna separada.

Company    = (Car_Dp['Company'])
Price      = (Car_Dp['Price'])
horsepower = (Car_Dh['horsepower'])

#concatentei as colunas em uma unica serie.

Cars = pd.Series()
Cars = pd.concat ([Company,Price,horsepower], axis = 1)

#Um DataFrame é simplesmente um conjunto de Series. 
#Transformei a serie em um data frame.

Cars_D = pd.DataFrame(Cars)
Cars_D
```

```python
# usando o index, troquei o index numerico per company
# usei o drop para remover a coluna company

Cars_D.index = Cars_D.Company
Cars_D.drop(['Company'], axis=1)
```

```python
#recriei o data frame a partir das series, adicionando uma nova coluna
#as notas foram atribuidas com base no preço.

Cars_D = pd.DataFrame(Cars)
Cars_D['Notas']=[7,5,9,8]
Cars_D
```

```python
url = 'https://raw.githubusercontent.com/pauloribeiroifpb/estatistica_aplicada/master/Automobile_data.csv'
Carros = pd.read_table(url,sep=',')
Carros
```

```python
#Função simples que percore cada coluna procurando um valor x e o substitui por Y.

Carros = Carros.apply(lambda x: x.replace('?', np.nan))
Carros = Carros.apply(lambda x: x.replace('n.a', np.nan))

Carros

```

```python
# usando o axis = 0, estou trabalhando as linhas
# how any seria, para qualquer ocorrencia nam dropar a linha

Carros = Carros.dropna( axis = 0, how = 'any')
```

```python
#para criar um dataframe com carros baratos e caros
#primeiro transformei o tipo em int32

Carros['price'] = Carros['price'].astype(int)

```

```python
Carros['price'].dtypes
```

```python
#criando dataframa com os carros que julguei como mais baratos

C_pobre = Carros.loc[(Carros['price'] < 10000)]
C_pobre
```

```python
#agora os mais caros
C_riquinho = Carros.loc[(Carros['price']>=30000)]
C_riquinho
```

```python
#carros de preço medio 

C_medio = Carros.loc[(Carros['price'] > 10000)]
C_medio = C_medio.loc[(C_medio['price'] <= 30000)]

C_medio
```
